```python

#当一个类存在两个独立变化的唯独，且这两个维度都需要进行拓展

#和抽象工厂很像，把变化维度拆分成两个抽象类，分别子类化拓展
#将一个类分成两个部分，
from abc import ABC,abstractmethod
class Messager:
    #第一部分：业务实现
    @abstractmethod
    def Login(self,username, password):
        pass
    @abstractmethod
    def  SendMessage(self,message):pass
    @abstractmethod
    def  SendPicture(self, image):pass

    #第二部分：平台实现
    @abstractmethod
    def  PlaySound(self):
        pass
    @abstractmethod
    def  DrawShape(self):
        pass
    @abstractmethod
    def WriteText(self):
        pass
    @abstractmethod
    def Connect(self):
        pass

class PCMessage(Messager):
    def  PlaySound(self):
        pass
    def  DrawShape(self):
        pass
  
    def WriteText(self):
        pass

    def Connect(self):
        pass

class PCMessageALL(PCMessage):

    def Login(self,username, password):
        pass

    def  SendMessage(self,message):pass

    def  SendPicture(self, image):pass
```

```python
# Implementor
class DrawingAPI:
    def draw_circle(self, x, y, radius):
        pass

# ConcreteImplementor 1/2
class DrawingAPI1(DrawingAPI):
    def draw_circle(self, x, y, radius):
        print(f"API1.circle at ({x}, {y}) radius {radius}")

# ConcreteImplementor 2/2
class DrawingAPI2(DrawingAPI):
    def draw_circle(self, x, y, radius):
        print(f"API2.circle at ({x}, {y}) radius {radius}")

# Abstraction
class Shape:
    def __init__(self, drawing_api):
        self.drawing_api = drawing_api
    
    def draw(self):
        pass
    
    def resize(self, radius):
        pass

# Refined Abstraction
class CircleShape(Shape):
    def __init__(self, x, y, radius, drawing_api):
        super().__init__(drawing_api)
        self.x = x
        self.y = y
        self.radius = radius
    
    def draw(self):
        self.drawing_api.draw_circle(self.x, self.y, self.radius)
    
    def resize(self, radius):
        self.radius = radius

# Client
class Client:
    def __init__(self):
        api1 = DrawingAPI1()
        api2 = DrawingAPI2()
        
        circle1 = CircleShape(1, 2, 3, api1)
        circle2 = CircleShape(4, 5, 6, api2)
        
        circle1.draw()
        circle2.draw()

        circle1.resize(7)
        circle2.resize(8)

        circle1.draw()
        circle2.draw()

client = Client()
```
